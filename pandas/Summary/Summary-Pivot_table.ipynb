```python
import numpy as np
import pandas as pd
```

## Example 1

```python
df = pd.DataFrame({'Name': ['Abby', 'Ben', 'Chulk', 'Danny', 'Ecko'],
                   'Gender': ['female', 'male', 'male', 'female', 'male'],
                   'Score': [85, 72, 86, 69, 90],
                   'Grade': ['B', 'C', 'B', 'D', 'A'],
                   'Pass Exam': [1, 1, 1, 0, 1]})
df
```

```python
df.pivot_table(values=['Score', 'Pass Exam'], index=['Gender'],
               aggfunc={'Score': [min, max, np.mean], 'Pass Exam':np.sum})
```

## Example 2

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
```

```python
df
```

### Aggregate values by taking the sum

```python
df.pivot_table(values='D', index=['A', 'B'], 
               columns=['C'], aggfunc=np.sum, fill_value=0)
```

### Aggregate by taking the mean across multiple columns

```python
df.pivot_table(values=['D', 'E'], index=['A', 'C'], 
               aggfunc={'D': np.mean, 'E': np.mean})
```

### Multiple types of aggregation for any given value column

```python
df.pivot_table(values=['D', 'E'], index=['A', 'C'], 
               aggfunc={'D': np.sum, 'E': [min, max, np.mean]}, 
               fill_value=0.0)
```
