```python
import numpy as np
import pandas as pd
```

```python
df_student = pd.DataFrame({'Student': ['Abby', 'Ben', 'Charlie', 'Daniel', 'Ecko'],
                           'Lecture': ['ML', 'DL', 'NLP', 'CV', 'ML']})
df_student
```

```python
df_lecture = pd.DataFrame({'Lecture': ['ML', 'DL', 'NLP', 'CV'], 
                           'Professor': ['Andrew', 'Goodfellow', 'Alex', 'LeCun']})
df_lecture
```

```python
df_nationality = pd.DataFrame({'Student': ['Abby', 'Ben', 'Charlie', 'Daniel', 'Ecko'],
                               'Nationality': ['USA', 'DE', 'FR', 'NL', 'CN']})
df_nationality
```

```python
df_lecture_language = pd.DataFrame({'Lecture': ['ML', 'DL', 'NLP', 'CV'], 
                                    'Language': ['Matlab', 'Python', 'Python', 'C++']})
df_lecture_language
```

```python
df_people_id = pd.DataFrame({'Name': ['Abby', 'Ben', 'Charlie', 'Daniel', 'Ecko', 'Andrew', 'Goodfellow', 'Alex', 'LeCun'],
                             'id': range(9)})
df_people_id
```

### One-to-one

```python
pd.merge(df_student,df_nationality)
```

### Many-to-one

```python
pd.merge(df_student, df_lecture)
```

### Many-to-many

```python
pd.merge(df_student, df_lecture_language)
```

### `on`

```python
pd.merge(df_student, df_lecture_language, on='Lecture')
```

```python
pd.merge(df_student, df_lecture, on='Lecture')
```

```python
pd.merge(df_student, df_nationality, on='Student')
```

### `left_on` and `right_on`

```python
pd.merge(df_student, df_people_id, left_on='Student', right_on='Name').drop('Name', axis=1)
```

### `left_index` and `right_index`

```python
df_student_indexed = df_student.set_index('Student')
df_student_indexed
```

```python
pd.merge(df_student_indexed, df_nationality, left_index=True, right_on='Student')
```

```python
df_lecture_student = df_student.set_index('Lecture')
df_lecture_student
```

```python
pd.merge(df_lecture, df_lecture_student, left_on='Lecture', right_index=True)
```

### Inner join

```python
pd.merge(df_student, df_people_id, left_on='Student', right_on='Name')
```

### Outer join

```python
pd.merge(df_student, df_people_id, left_on='Student', right_on='Name', how='outer')
```

### Left join

```python
pd.merge(df_student, df_people_id, left_on='Student', right_on='Name', how='left')
```

### Right join

```python
pd.merge(df_student, df_people_id, left_on='Student', right_on='Name', how='right')
```

```python
df_people_id_indexed = df_people_id.set_index('id')
df_people_id_indexed
```

```python
pd.merge(df_people_id_indexed, df_student, 
         left_on='Name', right_on='Student', how='left')
```
