### How to diplay plots

#### Plotting from a script

`plt.show()` starts an event loop, looks for all currently active figure objects, and opens one or more interactive windows that display figure or figures.

‼️ Note: 

The `plt.show()` command should be used **only once per Python session**, and is most often seen at the **very end of the script**.

````python
%run myplot.py
````

#### Plotting from an IPython shell

To enable this mode, use the `%matplotlib` magic command after starting `ipython`:

```python
In [1]: %matplotlib
Using matplotlib backend: TkAgg

In [2]: import matplotlib.pyplot as plt
```

At this point, any `plt` plot command will cause a figure window to open, and further commands can be run to update the plot. 

Some changes (such as modifying properties of lines that are already drawn) will not draw automatically: to force an update, use `plt.draw()`.

#### Plotting from an IPython notebook

Plotting interactively within an IPython notebook can be done with the `%matplotlib` command, and works in a similar way to the IPython shell. Two possible options of embedding graphics directly in the notebook:

- `%matplotlib notebook` will lead to *interactive* plots embedded within the notebook

- `%matplotlib inline` will lead to *static* images of your plot embedded in the notebook

````python
%matplotlib inline
````

After running this command (it needs to be done only once per kernel/session), any cell within the notebook that creates a plot will embed a PNG image of the resulting graphic:

````python
import numpy as np

x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), '-', label='$\sin(x)$')
plt.plot(x, np.cos(x), '--', label='$\cos(x)$')
plt.legend()
````

## Two Interfaces

A potentially confusing feature of Matplotlib is its dual interfaces: 

- convenient **MATLAB-style state-based** interface, and 

- a more powerful **object-oriented** interface.

### MATLAB-style Interface

The MATLAB-style tools are contained in the pyplot (`plt`) interface. 

````python
# create a plot figure
plt.figure()

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
````

This interface is *stateful*: it keeps track of the "current" figure and axes, which are where all `plt` commands are applied. You can get a reference to these using the `plt.gcf()` (get current figure) and `plt.gca()` (get current axes) routines.

### Object-oriented Interface (better way)

Rather than depending on some notion of an "active" figure or axes, in the object-oriented interface the plotting functions are *methods* of explicit `Figure` and `Axes` objects

Re-create the plots above:

````python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
````

### When to use which style?

- For more simple plots, the choice of which style to use is largely a matter of preference

- The object-oriented approach can become a necessity as plots become more complicated.
